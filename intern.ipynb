House Rented Data


```python
import pandas as pd
data=pd.read_csv('House_Rent_Dataset.csv')
data.head()
```

```python
y=data['Rent']
x=data.drop(['Rent','Posted On','Floor','Area Locality'],axis=1)
```

```python
import category_encoders as ce
encoder=ce.LeaveOneOutEncoder()
X =  encoder.fit_transform(x,y)
```

```python
print(X)
```

```python
#min max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(X)
```

```python
import pandas as pd 
x=pd.DataFrame(x,columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_train[1:2])
```

```python
new_house = pd.DataFrame({
    'BHK':[2],
    'Size':[1200],
    'Area Type':['Super Area'],
    'City':['Delhi'],
    'Furnishing Status':['Semi Furnished'],
    'Tenant Preferred':['Family'],
    'Bathroom':[2],
    'Point of Contact':['Contact Owner']
 

})
new_encoded = encoder.transform(new_house)
new_scaled = scaler.transform(new_encoded)
prediction=model.predict(new_scaled)
print("Predictedd Rent:",prediction[0])
```

```python
import joblib
artifacts = {
    "encoder":encoder,
    "scaler":scaler,
    "model":model
}
joblib.dump(artifacts,"house_rent_prediction.pkl")
```

```python
import pandas as pd

from sklearn.pipeline import Pipeline 
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score, mean_absolute_error 
import category_encoders as ce 
import joblib 

data = pd.read_csv("House_Rent_dataset.csv") 
data.head()
```

```python
y = data["Rent"] 
X = data.drop( 
columns=[ 
"Rent", 
"Posted On", 
"Floor", 
"Area Locality" 
] 
)
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
X, 
y, 
test_size=0.25, 
random_state=42 
)
```

```python
pipeline = Pipeline([ 
("encoder", ce.LeaveOneOutEncoder()), 
("scaler", MinMaxScaler()), 
("model", LinearRegression()) 
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
print("R² Score :", r2_score(y_test, y_pred)) 
print("MAE :", mean_absolute_error(y_test, y_pred)) 
```

```python
joblib.dump( 
pipeline, 
"house_rent_pipeline.pkl" 
)
```

```python
import pandas as pd
data=pd.read_csv("Lab  winequality-red.csv")
y=data['quality']
x=data.drop(['quality'],axis=1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=42)
model.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy of the model is: ",accuracy)
```
